# Weather data visualization using matplotlib

### Import dependencies

```python
# Import matplotlib for visualization
import matplotlib.pyplot as plt
# Import pandas to create dataframe
import pandas as pd
```

### Read weather data to dataframe

```python
# Read weather data from CSV to Pandas DataFrame
austin_weather = pd.read_csv('C:\\Users\\Robert.Jones\\Documents\\Python_Projects\\data_visualization\\weather_data\\austin_weather.csv', index_col ='DATE')
seattle_weather = pd.read_csv('C:\\Users\\Robert.Jones\\Documents\\Python_Projects\\data_visualization\\weather_data\\seattle_weather.csv', index_col = 'DATE')
```

### Explore the datasets

```python
# Brief explore of austin_weather dataset
austin_weather.style
```

```python
# Brief explore of seattle_weather dataset
seattle_weather.style
```

### Simple plot

```python
# Create a figure and axes
fig, ax = plt.subplots()

# Plot monthly precipitation from austin_weather against month (MLY-PRECP-NORMAL against MONTH)
ax.plot(austin_weather['MONTH'],austin_weather['MLY-PRCP-NORMAL'])

# Plot monthly precipitation from seattle_weather against month (MLY-PRECP-NORMAL against MONTH)
ax.plot(seattle_weather['MONTH'],seattle_weather['MLY-PRCP-NORMAL'])
```

### Show plot with custom line colors, markers, and linestyle

```python
# Create new instance of figure and axes
fig,ax = plt.subplots()

# Customize appearance of lines with color, markers, and linestyle

# Plot Seattle data
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"],color = 'b',marker = 'o', linestyle = '--')

# Plot Austin data
ax.plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"], color = 'r', marker = 'v', linestyle = '--')
```

### Show plot with axes labels and title

```python
# Create new instance of figure and axes
fig,ax = plt.subplots()

# Customizing axis labels and adding titles

# Plot Seattle data
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"],color = 'b',marker = 'o', linestyle = '--')

# Plot Austin data
ax.plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"], color = 'r', marker = 'v', linestyle = '--')

# Set x-axis label
ax.set_xlabel("Time (months)")

# Set y-axis label
ax.set_ylabel("Precipitation (inches)")

# Add the title
ax.set_title("Weather patterns in Austin and Seattle")
```

### Show multiple plots

```python
# Figure and axes for an array of subplots (2 rows, 2 columns)
fig,ax = plt.subplots(2,2)


# Adjust figure for visibility
fig.set_size_inches(12,8)
fig.tight_layout(pad=4.0)

### Plot the 4 charts
# Top left
ax[0, 0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"],color = 'b', linestyle = '--')
# Top right
ax[0, 1].plot(seattle_weather["MONTH"], seattle_weather["MLY-TAVG-NORMAL"],color = 'b', linestyle = '--')
# Bottom left
ax[1,0].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"],color = 'r', linestyle = '--')
ax[1,0].sharey(ax[0,0])
# Bottom right
ax[1,1].plot(austin_weather["MONTH"], austin_weather["MLY-TAVG-NORMAL"],color = 'r', linestyle = '--')
ax[1,1].sharey(ax[0,1])


# Set title and axes labes
ax[0,0].title.set_text("Seattle Precipitation by Months")
ax[0,0].set_xlabel("Time (months)")
ax[0,0].set_ylabel("Precipitation (inches)")

ax[0,1].title.set_text("Seattle Temperature by Months")
ax[0,1].set_xlabel("Time (months)")
ax[0,1].set_ylabel("Temperature (Celcius)")

ax[1,0].title.set_text("Austin Precipitation by Months")
ax[1,0].set_xlabel("Time (months)")
ax[1,0].set_ylabel("Precipitation (inches)")

ax[1,1].title.set_text("Austin Temperature by Months")
ax[1,1].set_xlabel("Time (months)")
ax[1,1].set_ylabel("Temperature (Celcius)")
```

### Small multiples with shared y-axis

```python
# Figure and axes for an array of subplots (2 rows, 1 columns),share y-axis
fig, ax = plt.subplots(2, 1, sharey=True);

# Adjust figure for visibility
fig.set_size_inches(12,8)
fig.tight_layout(pad=4.0)

### Plot Seattle precipitation between 25th and 75th percentile

# Plot Seattle precipitation data in the top axes
ax[0].plot(seattle_weather['MONTH'], seattle_weather['MLY-PRCP-NORMAL'], color = 'b');
ax[0].plot(seattle_weather['MONTH'], seattle_weather['MLY-PRCP-25PCTL'], color = 'b', linestyle = '--');
ax[0].plot(seattle_weather['MONTH'], seattle_weather['MLY-PRCP-75PCTL'], color = 'b', linestyle = '--');


### Plot Austin precipitation between 25th and 75th percentile

# Plot Austin precipitation data in the bottom axes
ax[1].plot(austin_weather['MONTH'], austin_weather['MLY-PRCP-NORMAL'], color = 'r');
ax[1].plot(austin_weather['MONTH'], austin_weather['MLY-PRCP-25PCTL'], color = 'r', linestyle = '--');
ax[1].plot(austin_weather['MONTH'], austin_weather['MLY-PRCP-75PCTL'], color = 'r', linestyle = '--');


### Set title and axes labes
# Set title and axes
ax[0].title.set_text("Seattle Precipitation Between 25th and 75th Percentile")
ax[1].title.set_text("Austin Precipitation Between 25th and 75th Percentile")

ax[0].set_xlabel("Time (months)")
ax[0].set_ylabel("Precipitation (inches)")

ax[1].set_xlabel("Time (months)")
ax[1].set_ylabel("Precipitation (inches)")
```
